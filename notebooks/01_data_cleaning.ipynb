```python
# Import Required Libraries
import pandas as pd
import numpy as np
```

```python
# Load Raw Dataset
df = pd.read_csv("../data/raw_data.csv")
df.head()
```

```python
## Basic Dataset Information

# Shape of dataset
print("Dataset Shape:", df.shape)

# Column info
df.info()


```

```python
#Checking Missing Values

df.isnull().sum()
```

```python

# Removing Duplicate Rows

df = df.drop_duplicates()

print("Duplicates after removal:", df.duplicated().sum())
print("New Shape:", df.shape)
```

```python

# Cleaning Rating Column


# Convert rating to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Fill missing ratings with median
df['rating'].fillna(df['rating'].median(), inplace=True)

df['rating'].describe()
```

```python

#  Cleaning Rating Count Column

# Remove text and keep only numbers
df['rating_count'] = df['rating_count'].str.extract('(\d+)')

# Convert to numeric
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Fill missing with 0
df['rating_count'].fillna(0, inplace=True)

df['rating_count'].head()
```

```python

#  Cleaning Cost Column


# Remove ₹ symbol and spaces
df['cost'] = df['cost'].str.replace('₹', '')
df['cost'] = df['cost'].str.replace(',', '')
df['cost'] = df['cost'].str.strip()

# Convert to numeric
df['cost'] = pd.to_numeric(df['cost'], errors='coerce')

# Fill missing with median
df['cost'].fillna(df['cost'].median(), inplace=True)

df['cost'].head()
```

```python

#  Handling Missing Values in Categorical Columns


df['city'].fillna("Unknown", inplace=True)
df['cuisine'].fillna("Unknown", inplace=True)
df['name'].fillna("Unknown", inplace=True)
```

```python

#  Final Missing Value Check


df.isnull().sum()
```

```python

# Fix Missing Address


df['address'].fillna("Address Not Available", inplace=True)

df.isnull().sum()
```

```python
#  Dropping Unnecessary Columns

df = df.drop(columns=['lic_no'])
```

```python

#  Save Final Cleaned Dataset


df.to_csv("../data/cleaned_data.csv", index=False)

print("Final cleaned dataset saved successfully!")
```
